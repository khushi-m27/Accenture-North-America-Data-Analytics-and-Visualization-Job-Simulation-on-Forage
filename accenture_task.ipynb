```python
# importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# reading csv files
content_df = pd.read_csv('/content/Content.csv')
reactions_df = pd.read_csv('/content/Reactions.csv')
reaction_types_df = pd.read_csv('/content/ReactionTypes.csv')
```

# **Data Cleaning**

```python
# finding rows that have missing values for all the tables
content_df.isna().sum() , reactions_df.isna().sum() , reaction_types_df.isna().sum()
```

Insight : There are few missing rows in tables content and reactions, hence we will remove the rows containing missing values.

```python
# content_df.dropna(inplace=True) # can be done but this will result in loss of data hence idea is to drop the URL column.
reactions_df.dropna(inplace=True)
reaction_types_df.dropna(inplace=True)
```

```python
#checking for any duplicated rows
content_df.duplicated().sum() , reactions_df.duplicated().sum() , reaction_types_df.duplicated().sum()
```

Insight :  No duplicated rows in all tables.

```python
#checking datatypes of each table
content_df.info()
```

```python
reactions_df.info()
```

```python
reaction_types_df.info()
```

```python
# droping irrelevant columns and correcting dtypes of each table columns
content_df.drop(['Unnamed: 0','User ID','URL'],axis=1,inplace=True)
reactions_df.drop(['Unnamed: 0','User ID'],axis=1,inplace=True)
reaction_types_df.drop('Unnamed: 0',axis=1,inplace=True)
reactions_df['Datetime'] = pd.to_datetime(reactions_df['Datetime'])
```

```python
# renaming type column in each table
content_df.rename(columns={'Type': 'Content Type'}, inplace=True)
reactions_df.rename(columns={'Type': 'Reaction Type'}, inplace=True)
reaction_types_df.rename(columns={'Type': 'Reaction Type'}, inplace=True)
```

```python
# checking value counts of each categorical column of each table.
#for content_df table
print("Value counts of Content table \n")
obj_columns = content_df.select_dtypes(include=['object']).columns
for column in obj_columns:
    print("Column:", column)
    print(content_df[column].value_counts())
    print("\n")
```

```python
# for reactions_df table
print("Value counts of Rections table \n")
obj_columns = reactions_df.select_dtypes(include=['object']).columns
for column in obj_columns:
    print("Column:", column)
    print(reactions_df[column].value_counts())
    print("\n")
```

```python
#for reaction_types_df table
print("Value counts of Rection types table \n")
obj_columns = reaction_types_df.select_dtypes(include=['object']).columns
for column in obj_columns:
    print("Column:", column)
    print(reaction_types_df[column].value_counts())
    print("\n")
```

Insight :  some categories have different format in content table

```python
# removing inverted commas
def remove_inverted_commas(value):
    return value.strip('"')

content_df['Category'] = content_df['Category'].apply(remove_inverted_commas)
```

```python
content_df['Category'] = content_df['Category'].str.lower() # converting all categories to lower case
content_df['Category'].unique() # cleaned column
```

```python
#list of columns in each table
content_df.columns.to_list(),reactions_df.columns.to_list(),reaction_types_df.columns.to_list()
```

```python
#merging each table into one table
merged_df = pd.merge(reactions_df, content_df, on='Content ID', how='left')
merged_df = pd.merge(merged_df, reaction_types_df, on='Reaction Type', how='left')
merged_df.head()
```

```python
category_scores = merged_df.groupby('Category')['Score'].sum().reset_index()
category_scores_sorted = category_scores.sort_values(by='Score', ascending=False)
top_five_categories = category_scores_sorted.head(5)
```

```python
top_five_categories
```

```python
with pd.ExcelWriter('social_buzz_analysis.xlsx') as writer:
    merged_df.to_excel(writer, sheet_name='Cleaned Data', index=False)
    top_five_categories.to_excel(writer, sheet_name='Top Five Categories', index=False)
```

# Data Visualisation

```python
#bar chart for top five categories
plt.figure(figsize=(10, 6))
sns.barplot(x='Score', y='Category', data=top_five_categories, palette='viridis')
plt.title('Top 5 Content Categories by Total Score')
plt.xlabel('Total Score')
plt.ylabel('Category')
plt.savefig('top_5_categories.png', format='png')
plt.show()
```

```python
#no. of unique categories
unique_categories = merged_df['Category'].nunique()
print("Number of unique categories:", unique_categories)
```

```python
# reactions from the most popular category
most_popular_category = top_five_categories.iloc[0]['Category']  # top category
reactions_most_popular = merged_df[merged_df['Category'] == most_popular_category]['Reaction Type'].count()
print(f"Number of reactions to the most popular category ({most_popular_category}):", reactions_most_popular)
```

```python
#month with most posts
merged_df['Month'] = merged_df['Datetime'].dt.strftime('%b')
posts_per_month = merged_df.groupby('Month')['Content ID'].count().reset_index()
posts_per_month.rename(columns={'Content ID': 'Total_posts'}, inplace=True)
month_order = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
posts_per_month['Month'] = pd.Categorical(posts_per_month['Month'], categories=month_order, ordered=True)
most_posts_month = posts_per_month.sort_values(by='Total_posts', ascending=False).iloc[0]
print("Month with the most posts:", most_posts_month['Month'])
print("Number of posts:", most_posts_month['Total_posts'])


plt.figure(figsize=(12, 6))
sns.lineplot(x='Month', y='Total_posts', data=posts_per_month, marker='o', color='b')
plt.title('Number of Posts per Month')
plt.xlabel('Month')
plt.ylabel('Number of Posts')
plt.savefig('month_with_most_posts.png', format='png')
plt.show()
```
